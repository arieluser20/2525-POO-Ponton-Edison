```python
import tkinter as tk
from tkinter import messagebox, filedialog

class VerificadorApp:
    def __init__(self, root):
        self.root = root
        root.title("Verificador Lógico")
        root.geometry("400x350")
        root.resizable(False, False)
        self.historial = []

        tk.Label(root, text="Ingrese un número:", font=("Arial", 13)).pack(pady=10)
        self.entry_num = tk.Entry(root, font=("Arial", 13), width=15, justify="center")
        self.entry_num.pack(pady=5)

        tk.Button(root, text="Verificar", command=self.verificar, width=15, bg="#3498db", fg="white").pack(pady=10)
        self.result_var = tk.StringVar()
        tk.Label(root, textvariable=self.result_var, font=("Arial", 14, "bold"), fg="#2ecc71").pack(pady=5)

        tk.Label(root, text="Historial:", font=("Arial", 12)).pack(pady=10)
        self.listbox = tk.Listbox(root, width=30, height=8)
        self.listbox.pack(pady=5)

        tk.Button(root, text="Limpiar", command=self.limpiar, width=12, bg="#e74c3c", fg="white").pack(pady=5)
        tk.Button(root, text="Guardar Historial", command=self.guardar_historial, width=15, bg="#f1c40f", fg="black").pack(pady=5)

    def verificar(self):
        valor = self.entry_num.get().strip()
        if not valor.isdigit():
            messagebox.showerror("Error", "Ingrese un número entero válido.")
            return
        num = int(valor)
        if num >= 15:
            resultado = "Sí"
        elif num <= 7:
            resultado = "No"
        else:
            resultado = "Indeterminado"
        self.result_var.set(f"Resultado: {resultado}")
        self.historial.append(f"{num} → {resultado}")
        self.actualizar_historial()

    def actualizar_historial(self):
        self.listbox.delete(0, tk.END)
        for item in self.historial:
            self.listbox.insert(tk.END, item)

    def limpiar(self):
        self.entry_num.delete(0, tk.END)
        self.result_var.set("")
        self.historial.clear()
        self.actualizar_historial()

    def guardar_historial(self):
        archivo = filedialog.asksaveasfilename(defaultextension=".txt",
                                               filetypes=[("Archivo de texto", "*.txt")])
        if archivo:
            with open(archivo, "w") as f:
                for item in self.historial:
                    f.write(item + "\n")
            messagebox.showinfo("Guardado", "Historial guardado correctamente.")

if __name__ == "__main__":
    root = tk.Tk()
    app = VerificadorApp(root)
    root.mainloop()
```
